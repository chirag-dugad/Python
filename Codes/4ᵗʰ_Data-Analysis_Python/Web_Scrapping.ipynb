```python
import requests # pip install requests
from bs4 import BeautifulSoup # pip install beautifulsoup4
import pandas as pd
```

# 1. Website(Table) to .csv
We extract all the table from the website

```python
def extract_tables_to_csv(url):
    """Extracts tables from a given URL and saves them to CSV files.

    Args:
        url: The URL of the webpage to scrape.
    """

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    tables = soup.find_all('table')

    for i, table in enumerate(tables):
        # Convert table to Pandas DataFrame
        df = pd.read_html(str(table))[0]

        # Save DataFrame to CSV
        filename = f"table_{i+1}.csv"
        df.to_csv(filename, index=False)
        print(f"Table {i+1} saved to {filename}")

# Replace with the target URL
url = "https://www.ascii-code.com/"  # Replace with the actual URL
extract_tables_to_csv(url)
```

# 2. Website(Image) to .jpg
Extract all the Images from the Website

```python
def download_images(url):
  """Downloads all images from a given URL.

  Args:
    url: The URL of the webpage to scrape.
  """

  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')

  images = soup.find_all('img')

  for img in images:
    img_url = img['src']

    # Handle relative URLs
    if not img_url.startswith('http'):
      img_url = f"{url}/{img_url}"

    # Download the image
    image_response = requests.get(img_url)

    # Create a filename based on the image URL
    filename = img_url.split('/')[-1]

    with open(filename, 'wb') as f:
      f.write(image_response.content)

    print(f"Downloaded {filename}")

# Replace with the target URL
url = "https://www.hopscotch.in/blog/5-famous-akbar-birbal-moral-stories/"

download_images(url)
```
